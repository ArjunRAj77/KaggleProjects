Let's look at code of each of these visualizations.

### Barplot()

As the name suggests, this function is used for creating bar chart from a given set of data. 



Let's start with inputting dataset to the notebook.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
museum_filepath = "../input/data-for-datavis/museum_visitors.csv"
flight_filepath = "../input/data-for-datavis/flight_delays.csv"
insurance_filepath = "../input/data-for-datavis/insurance.csv"

# Fill in the line below to read the file into a variable museum_data
museum_data = pd.read_csv(museum_filepath,index_col="Date",parse_dates=True)
flight_data=pd.read_csv(flight_filepath,index_col="Month")
insurace_data=pd.read_csv(insurance_filepath)

# Displaying the top 5 values of data
flight_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

We can see that the month June (6th) have the most delays in flight.

- `sns.barplot` tells the notebook that we want a bar chart.
- `x=flight_axis` denotes or control the x axis
- ` y = flight_data['NK']` denotes the y axis

## Heatmaps

- we create a heatmap to quickly visualize patterns.
- Each cell is color-coded according to its corresponding value.


```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
```

## Scatterplot

To create a simple scatter plot, we use the `sns.scatterplot` command and specify the values for:
- the horizontal x-axis 
- the vertical y-axis 

```python

sns.scatterplot(x=insurace_data['bmi'], y=insurace_data['charges'])
```

>The scatterplot above suggests that body mass index (BMI) and insurance charges are positively correlated, where customers with higher BMI typically also tend to pay > more in insurance costs. (This pattern makes sense, since high BMI is typically associated with higher risk of chronic disease.) 
>A regression line can be added by:

```python
# color-coding scatter plots
sns.scatterplot(x=insurace_data['bmi'], y=insurace_data['charges'], hue=insurace_data['smoker'])
```

### Regplot()
This is similar to scatterplot. Unlike scatterplot, a regression line will be made available with the plot.

```python
sns.regplot(x=insurace_data['bmi'], y=insurace_data['charges'])
```

### Lmplot()
It is used too add two regression lines, corresponding to variables.

```python

sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurace_data)
```

### Swarmplot


```python

plt.figure(figsize=(20,7))

#Using lmplot
sns.swarmplot(x=insurace_data['smoker'],y=insurace_data['charges'])
```

## Distribution visualization techniques



### histplot()
This function is used to create a histogram from data.


```python
iris_path="../input/data-for-datavis/iris.csv"
iris_data=pd.read_csv(iris_path)
sns.histplot(iris_data['Petal Length (cm)'])
```

### kdeplot()
- kernal density plot
- kind of smoothed histogram

To make a KDE plot, we use the `sns.kdeplot` command. Setting `shade=True `colors the area below the curve (and data= chooses the column we would like to plot).

```python
# KDE plot 
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

### jointplot()

- two-dimensional (2D) KDE plot


```python
# 2D KDE plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```
